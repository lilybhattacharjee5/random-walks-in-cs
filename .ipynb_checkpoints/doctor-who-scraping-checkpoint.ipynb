```python
class InteractionNode:
    def __init__(self, name, id_num):
        self.name = name
        self.adjacents = {}
        self.id = id_num
        
    def addAdj(self, adj, sent):
        if adj in self.adjacents.keys():
            self.adjacents[adj][0] += 1
            self.adjacents[adj][1][0] += sent['neg'] # red ~ negative
            self.adjacents[adj][1][1] += 0 # green ~ neutral
            self.adjacents[adj][1][2] += sent['pos'] # blue ~ positive
        else:
            self.adjacents[adj] = [0, [0, 0, 0]]
            self.adjacents[adj][0] = 1
            self.adjacents[adj][1][0] = sent['neg'] # red
            self.adjacents[adj][1][1] = 0 # green
            self.adjacents[adj][1][2] = sent['pos'] # blue
        self.recalculate_rgb(adj)
            
    def recalculate_rgb(self, adj):
        rgb_setup = self.adjacents[adj][1]
        overall_rgb = sum(self.adjacents[adj][1])
        if overall_rgb == 0:
            overall_rgb = 1
        rgb_setup[0] = rgb_setup[0] / overall_rgb * 256
        rgb_setup[1] = rgb_setup[1] / overall_rgb * 256
        rgb_setup[2] = rgb_setup[2] / overall_rgb * 256

class InteractionGraph:
    def __init__(self):
        self.nodes = set()
        self.node_names = {}
        
    def addNode(self, node):
        self.nodes.add(node)
        self.node_names[node.name] = node
        
    def removeNode(self, node):
        self.nodes.remove(node)
        del self.node_names[node.name]
```

```python
# generalize into utility functions
def access_url(doctor, episode):
    urls = ["http://www.chakoteya.net/DoctorWho/" + str(doctor) + "-" + str(episode) + ".html",
           "http://www.chakoteya.net/DoctorWho/" + str(doctor) + "-" + str(episode) + ".htm"]
    print(urls[0])
    page_response = requests.get(urls[0], timeout = 5)
    if page_response.status_code == 404:
        print(urls[1])
        page_response = requests.get(urls[1], timeout = 5)
    return page_response

def get_content(page_response):
    page_content = BeautifulSoup(page_response.content, "html.parser")
    textContent = []
    paragraphs = page_content.find_all("p")
    for i in range(len(paragraphs)):
        textContent.append(re.sub(r"[\r\n]+", r"", paragraphs[i].text))
    return textContent
    
def sep_chars(textContent):
    elem_matches = []
    sent_lst = []
    for elem in textContent:
        if type(elem) != str:
            continue
        matches = list(set(map(drop_colon, re.findall(r"[A-Z]+:", elem))))
        if len(matches) > 0:
            elem_matches.append(matches)
            try:
                sent_lst.append(sentiment_analyzer_scores(elem))
            except:
                print("")
    return [elem_matches, sent_lst]
    
def gen_interaction_graph(elem_matches, sent_lst):
    ep = InteractionGraph()
    id_num = 0
    for interaction_idx in range(len(elem_matches)):
        elem_matches[interaction_idx] = [i for i in elem_matches[interaction_idx] if i != 'ALL']
    ctr = 0
    for interaction in elem_matches:
        if ctr >= len(sent_lst):
            curr_sent = {"pos": 0, "neg": 0, "neu": 0}
        else:
            curr_sent = sent_lst[ctr]
        for char in interaction:
            if char not in ep.node_names:
                # edit the existing node
                new_node = InteractionNode(char, id_num)
                id_num += 1
                ep.addNode(new_node)
            old_node = ep.node_names[char]
            spec_interaction = interaction[:]
            spec_interaction.remove(char)
            if spec_interaction != None:
                for adj in spec_interaction:
                    if adj in ep.node_names.keys():
                        old_node.addAdj(adj, curr_sent)
                    else:
                        new_adj = InteractionNode(adj, id_num)
                        id_num += 1
                        ep.addNode(new_adj)
                        old_node.addAdj(adj, curr_sent)
            else:
                if ep.node_names[char] == None:
                    # create a new node
                    new_node = InteractionNode(char, id_num)
                    id_num += 1
                    # add the node to the interaction graph
                    ep.addNode(new_node)
        ctr += 1
    return ep
```

```python
# let's generalize this to every doctor in every episode
def doctor_graph(ep_range, season_range):
    page_content = []
    for season in season_range:
        for ep in ep_range[str(season)]:
            try:
                page_response = access_url(season, ep)
                page_content.extend(get_content(page_response))
                time.sleep(5)
            except:
                print("except")
                continue
    elem_matches, sent_lst = sep_chars(page_content)
    ep_graph = gen_interaction_graph(elem_matches, sent_lst)
    return ep_graph
```

```python
# generalize to all doctor groups
def create_node_json(doctor_graph):
    doctor_nodes = []
    for char_name, char_node in doctor_graph.node_names.items():
        doctor_nodes.append({
            "character": char_name,
            "id": char_node.id,
            "influence": len(char_node.adjacents),
        })
    return doctor_nodes

def create_link_json(doctor_graph):
    doctor_links = []
    for char_name, char_node in doctor_graph.node_names.items():
        char_adj = list(char_node.adjacents)
        for adj in char_adj:
            target_node = doctor_graph.node_names[adj]
            doctor_links.append({
                "source": char_node.id,
                "target": target_node.id,
                "weight": char_node.adjacents[target_node.name][0],
                "red": char_node.adjacents[target_node.name][1][0],
                "green": char_node.adjacents[target_node.name][1][1],
                "blue": char_node.adjacents[target_node.name][1][2]
            })
    return doctor_links
    
def create_data_file(doctor_filename, doctor_nodes, doctor_links):
    doctor_file = open(doctor_filename, "w+")
    doctor_file.write("{\"nodes\": ")
    doctor_file.write(str(doctor_nodes).replace("'", "\""))
    doctor_file.write(", \"links\": ")
    doctor_file.write(str(doctor_links).replace("'", "\""))
    doctor_file.write("}")
    print(doctor_filename)
    doctor_file.close()
```
